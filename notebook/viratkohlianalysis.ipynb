```python
deliveries = pd.read_csv("../data/raw/deliveries.csv")
matches = pd.read_csv("../data/raw/matches.csv")
```

```python
virat_df = deliveries[deliveries['batsman'] == 'V Kohli']
```

```python
virat_df.head()
```

```python
virat_df.shape
```

```python
virat_df = virat_df.merge(
    matches[['id', 'season']],
    left_on='match_id',
    right_on='id',
    how='left'
)
```

```python
virat_df[["match_id","season"]].head()
```

```python
total_runs = virat_df['batsman_runs'].sum()
total_runs
```

```python
total_balls = virat_df.shape[0]
total_balls
```

```python
strike_rate = (total_runs / total_balls) * 100
strike_rate
```

```python
season_runs = (
    virat_df
    .groupby('season')['batsman_runs']
    .sum()
    .reset_index()
)
season_runs
```

```python
season_balls = (
    virat_df
    .groupby('season')
    .size()
    .reset_index(name='balls')
)
season_balls
```

```python
season_stats1 = season_runs.merge(season_balls, on='season')
season_stats1
```

```python
season_stats1['strike_rate'] = (
    season_stats1['batsman_runs'] / season_stats1['balls']
) * 100

season_stats1
```

```python
season_stats1.head()
```

```python
match_runs = (
    virat_df
    .groupby('match_id')['batsman_runs']
    .sum()
    .reset_index()
)

match_runs.head()
```

```python
matches_played = match_runs.shape[0]
matches_played
```

```python
avg_runs = match_runs['batsman_runs'].mean()
avg_runs
```

```python
thirty_plus = match_runs[match_runs['batsman_runs'] >= 30].shape[0]
thirty_plus
```

```python
consistency_pct = (thirty_plus / matches_played) * 100
(consistency_pct)
```
